# Web-Scraping

```python
#Now we try to do the static web-scrape, it used to work and produce a DataFrame of 12 cars, but it doesnt now for some reason
#It says Access Denied now, proper gutted, smh.

url = 'https://www.autotrader.co.uk/car-search?postcode=G72%208JD&year-to=2022&include-delivery-option=on&advertising-location=at_cars&page=1'
response = requests.get(url)
print(response)

soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify()) 
```

```python
#Create Empty lists to hold the data
name = []
price = []
year = []
body = []
miles = []
engine_volume = []
engine_power =[]
transmission = []
fuel = []

#Scrapping Time

name_results  = soup.find_all('h3', class_='product-card-details__title')
for i in range(len(name_results)):
    name_stripped = name_results[i].text.strip()
    #print(name_results[i].text.strip())
    name.append(name_stripped)
price_results = soup.find_all('div', class_='product-card-pricing__price')
for i in range(len(price_results)):
    price_stripped = price_results[i].text.strip()
    #print(price[i].text.strip())
    price.append(price_stripped)
specs1 = soup.find_all('ul', attrs={'class': 'listing-key-specs'})
for i in specs1:
    spec_stripped = i.text.strip()
    spec_splitted = spec_stripped.split("\n")
    year.append(spec_splitted[0])
    body.append(spec_splitted[1])
    miles.append(spec_splitted[2])
    engine_volume.append(spec_splitted[3])
    engine_power.append(spec_splitted[4])
    transmission.append(spec_splitted[5])
    fuel.append(spec_splitted[6])
    print("-------")
    print(name)
    print(price)
    print(year)
    print(body)
    print(miles)
    print(engine_volume)
    print(engine_power)
    print(transmission)
    print(fuel)
    print("-------")
    
```

```python
#Zipping our lists into a DataFrame
zipped =list(zip(name, price, year,body,miles,engine_volume,engine_power,transmission,fuel))


df = pd.DataFrame(zipped, columns=['Model', 'Price', 'Year',"Body","Mileage","Engine_volume","Engine_power","Transmission","Fuel"])
df
```
