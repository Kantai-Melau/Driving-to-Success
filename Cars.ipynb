## Import and Process Data

```python
file_path = "Resources/vehicles.csv"
cars_df = pd.read_csv(file_path)
cars_df.head()
```

```python
#Check the existing columns
cars_df.columns
```

```python
#Drop unwanted columns
new_cars_df = cars_df[["price", "year", "manufacturer", "model", "odometer", "fuel", "transmission"]]
new_cars_df
```

```python
new_cars_df = new_cars_df.dropna(how ='any')
#new_cars_df = new_cars_df.reset_index()
new_cars_df
```

```python
new_cars_df = new_cars_df.reset_index()
new_cars_df = new_cars_df[['price', 'year', 'manufacturer', 'model', 'odometer', 'fuel', 'transmission']]
new_cars_df
```

```python
new_cars_df = new_cars_df.rename(columns = {'price': 'Price',
                                            'year': 'Year',
                                            'manufacturer': 'Manufacturer',
                                            'model': 'Model',
                                            'odometer': 'Mileage',
                                            'fuel': 'Fuel',
                                            'transmission': 'Transmission'
                                            }
                                )
new_cars_df
```
